```python
import pandas as pd
import geopandas as gpd

data = pd.read_csv('covid_cases.csv')
    
data.head()
```

```python
data_cases = data[['Country/Region','Confirmed']]


world_data = gpd.read_file('corona_cases\\World_Map.shp')

for items in data_cases['Country/Region'].tolist():
    world_data_list = world_data['NAME'].tolist()
    if items in world_data_list:
        pass
    else:
        print(items + ' is not in the world_data list')
```

```python
world_data['NAME'].replace('Brunei Darussalam', 'Brunei', inplace = True)
world_data['NAME'].replace('Cape Verde', 'Cabo Verde', inplace = True)
world_data['NAME'].replace('Congo', 'Congo (Kinshasa)', inplace = True)
world_data['NAME'].replace('Democratic Republic of the Congo', 'Congo (Brazzaville)', inplace = True)
world_data['NAME'].replace('Czech Republic', 'Czechia', inplace = True)
world_data['NAME'].replace('Swaziland', 'Eswatini', inplace = True)
world_data['NAME'].replace('United States', 'US', inplace = True)
world_data['NAME'].replace('Viet Nam', 'Vietnam', inplace = True)
world_data['NAME'].replace('Iran (Islamic Republic of)', 'Iran', inplace = True)
world_data['NAME'].replace('The former Yugoslav Republic of Macedonia', 'North Macedonia', inplace = True)
world_data['NAME'].replace("Korea, Democratic People's Republic of", 'South Korea', inplace = True)
world_data['NAME'].replace("Holy See (Vatican City)", 'Holy See', inplace = True)
world_data['NAME'].replace("Lao People's Democratic Republic", 'Laos', inplace = True)
world_data['NAME'].replace("Libyan Arab Jamahiriya", 'Libya', inplace = True)
world_data['NAME'].replace("Republic of Moldova", 'Moldova', inplace = True)
world_data['NAME'].replace("Syrian Arab Republic", 'Syria', inplace = True)
world_data['NAME'].replace("Taiwan", 'Taiwan*', inplace = True)
world_data['NAME'].replace("United Republic of Tanzania", 'Tanzania', inplace = True)
world_data['NAME'].replace("United Republic of Tanzania", 'Tanzania', inplace = True)
world_data['NAME'].replace("Sudan", 'South Sudan', inplace = True)
world_data['NAME'].replace("Palestine", 'West Bank and Gaza', inplace = True)
```

```python
data_cases.rename(columns = {'Country/Region': 'NAME'}, inplace = True)

combined = world_data.merge(data_cases, on = 'NAME')

combined.to_file(r'corona_cases\\combined.shp')
```
